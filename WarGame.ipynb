```python
import random
```

```python
suits = ['diamond','heart','spade','club']
values = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9, 
        'ten':10, 'eleven':11, 'twelve':12, 'thirteen':13}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    
    def __str__(self):
        return f'{self.rank} of {self.suit}'
        
```

```python
class Deck:
    
    def __init__(self):
        self.all_cards = []
        
        for x in suits:
            for y in values:
                self.all_cards.append(Card(x,y))
    
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def del_one(self):
        return self.all_cards.pop(0)
        
    def __str__(self):
        return str(len(self.all_cards))
    
    def print_all(self):
        return self.all_cards
```

```python
class Player:
    
    def __init__(self,name):
        self.name = name
        self.all_cards = []
    
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def add_cards(self,new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return f'{self.name} has {len(self.all_cards)} card'
        
```

```python
# the war game
```

```python
player1 = Player('Ashish')
player2 = Player('Abha')

deck = Deck()
deck.shuffle()

for x in range(0,26):
    player1.add_cards(deck.del_one())
    player2.add_cards(deck.del_one())

game_on = True
round = 0
while game_on:
    
    round +=1
    print(f'round {round} begins!')
    
    if len(player1.all_cards) == 0:
        print(f'{player1.name} lost the game! {player2.name} Won!!')
        game_on = False
        break
        
    if len(player2.all_cards) == 0:
        print(f'{player2.name} lost the game! {player1.name} Won!!')
        game_on = False
        break
    
    player1_hand_card = []
    player1_hand_card.append(player1.remove_one())
    player2_hand_card = []
    player2_hand_card.append(player2.remove_one())
    
    war = True
    
    while war:
        
        if  player1_hand_card[-1].value > player2_hand_card[-1].value:
            player1.add_cards(player1_hand_card)
            player1.add_cards(player2_hand_card)
            war = False
            break
            
        elif player1_hand_card[-1].value < player2_hand_card[-1].value:
            player2.add_cards(player1_hand_card)
            player2.add_cards(player2_hand_card)
            war = False
            break
        
        else:
            #warzone
            print("warzone begins")
            
            if len(player1.all_cards) < 5:
                print(f'{player1.name} failed the warzone! {player2.name} won!!')
                game_on = False
                break
                
            elif len(player2.all_cards) < 5:
                print(f'{player2.name} failed the warzone! {player1.name} won!!')
                game_on = False
                break
                
            else:
                for x in range(0,5):
                    player1_hand_card.append(player1.remove_one())
                    player2_hand_card.append(player2.remove_one())
                
                
            
```
